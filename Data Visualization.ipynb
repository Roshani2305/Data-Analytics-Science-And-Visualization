```python
import matplotlib.pyplot as mpl

companies=['MS','Google','TCS','HP','Wipro','Infosys','Dell','GVOS','HCL','Mahindra']
jobs=[700,750,455,345,350,538,672,363,547,265]
clr=['red','blue','green','magenta','cyan','orange','purple','black','yellow','grey']

mpl.bar(companies,jobs,color=clr,width=0.50,bottom=100)
mpl.xlabel('Companies')
mpl.ylabel('No. of Jobs')
mpl.title('Jobs in 2021')

mpl.show()
```

```python
import matplotlib.pyplot as mpl

Company=['Microsoft','Google','TCS','HP','Wipro','Infosys','Dell','GVOS','HCL','Mahindra']
Jobs=[700,750,455,345,350,538,672,363,547,265]

mpl.pie(Jobs,labels=Company)
mpl.title('Jobs Available in 2021')
mpl.show()
```

```python
from matplotlib import pyplot as mpl
%matplotlib inline

Company=['MS','Google','TCS','HP','Wipro','Infosys','Dell','GVOS','HCL','Mahindra']
Jobs=[700,750,455,345,350,538,672,363,547,265]
AvgSalary=[409992,319666,374142,420000,342363,286112,299996,126622,221400,329945]

fig, axes=mpl.subplots(1,2)
axes[0].plot(Company,Jobs,'g--')

axes[1].plot(Company,AvgSalary,'r--')
mpl.show()
```
